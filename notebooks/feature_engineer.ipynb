```python
def to_interim(df, mode, save=True):
    
    # make sure mode arg is valid then load up data from raw folder
    assert mode in ['train', 'test']
    
    df = pd.read_csv(mapper.raw_data_dir/f'{mode}.csv')
    
    # break out date cols, dropping date fields when finished
    df = handle_date_cols(df, ['first_active_month'], add_date_parts=True, drop_date_cols=True)
    
    # want target at the end of train df so move things around a bit
    if mode == 'train':
        df_columns_reordered = list(df.columns.drop('target')) + ['target']

        df = df[df_columns_reordered]
    
    # save out and return 
    df.to_parquet(mapper.interim_data_dir/f'{mode}.parquet')
    
    return df
```
